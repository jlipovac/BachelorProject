```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=["correct by\nname", "correct by\ntax id", "wrong by\nname", "wrong by\ntax id", "unclassified"]
big=[50.42, 56.22, 38.83, 33.03, 10.75]
small1=[49.84, 55.68, 39.35, 33.51, 10.81]
small2=[49.84, 56.24, 40.23, 33.83, 9.93]
line1, = plt.plot(x,big,label='Big data', linestyle='-', marker='.')
line2, = plt.plot(x,small1,label="Sara's small data", linestyle='--')
line3, = plt.plot(x,small2,label="Tea's small data", linestyle=':')
plt.title("Classification results")
plt.ylabel('%', fontsize=8)
plt.legend(handles=[line1, line2, line3], loc='upper right')
plt.grid(visible=True, which='major', color='black', alpha=0.6, linestyle='-', lw=0.6)
plt.minorticks_on()
plt.grid(visible=True, which='minor', color='gray', alpha=0.5, ls=':', lw=0.5)
plt.show()
```

```python
df = pd.DataFrame({
    '': ["correct by\nname", "correct by\ntax id", "wrong by\nname", "wrong by\ntax id", "unclassified"],
    'Big data': [50.42, 56.22, 38.83, 33.03, 10.75],
    "Sara's small data": [49.84, 55.68, 39.35, 33.51, 10.81],
    "Tea's small data": [49.84, 56.24, 40.23, 33.83, 9.93]
})

ax = df.plot.bar(x="", y=["Big data", "Sara's small data", "Tea's small data"], title="Classification results", rot=90)
ax.set_axisbelow(True)
ax.grid(zorder=0)
```

```python
labels = ["correct by name", "wrong by name", "unclassified"]
colors = ["violet", '#9932CC', "#DDA0DD"]
myexplode=[0.1, 0.1, 0]
sizes = [50.42, 38.83, 10.75]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, explode=myexplode, shadow=True)
ax.axis('equal')
ax.set_title('Classification on big data\n')
plt.show()
```
